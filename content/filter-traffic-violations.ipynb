## Filter the traffic violations file

- Keep only every tenth line.
- Delete lines that don't have a full set of columns.


```python
openFile = open('Traffic_Violations.csv')

f2 = open('Traffic_Violations_filtered.csv', 'w')

for ct, line in enumerate(openFile, 0):
    if ct % 10:
        continue
        
    columns = line.split('|')
    if len(columns) < 35:
        print('Short column at line', ct)
        print(columns)
        continue
    f2.write(line)
    
openFile.close()
f2.close()
```
